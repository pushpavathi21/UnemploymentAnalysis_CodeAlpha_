```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
df.head()

```

```python
# Check column names and data types
df.info()

```

```python
# Check for null values
df.isnull().sum()
```

```python
# Descriptive statistics
df.describe()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.replace('.', '', regex=False).str.replace(' ', '_')

# View updated column names
df.columns

```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Estimated_Unemployment_Rate_(%)', data=df)
plt.xticks(rotation=90)
plt.title('Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
# Convert Date column to datetime if not already done
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(14, 6))
sns.lineplot(x='Date', y='Estimated_Unemployment_Rate_(%)', hue='Region', data=df)
plt.title('Unemployment Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Estimated_Employed', y='Estimated_Labour_Participation_Rate_(%)', hue='Region', data=df)
plt.title('Employment vs Labour Participation Rate')
plt.xlabel('Estimated Employed')
plt.ylabel('Labour Participation Rate (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
